**ðŸ§© Filtering Rows (Simple condition)**

```python
# Filter the DataFrame to include only rows where the 'Age' column value is greater than 30.
# This creates a new DataFrame containing a subset of the original data.
# The condition (df_pb['Age'] > 30) returns a boolean Series (True/False) for each row.
df_pb[df_pb['Age'] > 30]
```

**ðŸ§© Filtering Rows (Multiple condition)**

```python
# Filter the DataFrame to include only rows where BOTH of these conditions are true:
# 1. The 'Age' column value is greater than 25
# 2. The 'City' column value is exactly "Delhi"

# The '&' operator is used to combine multiple conditions (logical AND). 
# Each condition must be enclosed in parentheses due to operator precedence.
# The result is a subset DataFrame containing only the rows that satisfy both criteria.
df_pb[(df_pb['Age'] > 25) & (df_pb['City'] == "Delhi")]
```

**OR condition**

```python
# Select rows where Age > 30 OR City is Mumbai
df_pb[(df_pb["Age"] > 30) | (df_pb["City"] == "Mumbai")]
```

**NOT condition**

```python
# Select rows where City is NOT Kolkata
df_pb[~(df_pb['City'] == "Kolkata")]

# Select rows where City is NOT Kolkata (Method 2)
df_pb[df_pb['City'] != "Kolkata"]
```

**Nested conditions**

```python

# Select rows where (Age > 25 AND City is Delhi) OR (Salary > 50000), and sort by salary_INR
df_pb[((df_pb['Age'] > 25) & (df_pb['City'] == "Delhi")) | (df_pb['Salary_INR'] > 50000)].sort_values(by='Salary_INR',ascending=False)
```

**Combination of AND, OR, NOT**

```python
# Select rows where Age > 30 AND (City is NOT Mumbai OR Salary > 70000)
df_pb[(df_pb["Age"] >30) & ((df_pb["City"] != "Mumbai") | (df_pb['Salary_INR'] > 70000))].sort_values(by='Salary_INR', ascending=False)
```

**Multiple nested conditions**

```python
# Age between 25 and 35, City Delhi or Mumbai, and Salary > 40000
df_pb[((df_pb['Age'] >= 25) & (df_pb['Age'] <= 35)) \
    & ((df_pb['City'] == "Delhi") \
    | (df_pb['City'] == "Mumbai")) \
        & (df_pb['Salary_INR'] > 40000)]
```

```python
# Select rows where Salary >= 50000 OR Age < 30
df_pb[(df_pb['Salary_INR'] >= 50000) | (df_pb['Age'] < 30)]
```

**ðŸŸ  Intermediate Exercises**

```python
# (Age > 25 AND City is Delhi) OR (Salary > 70000)
df_pb[((df_pb['Age'] > 25) & (df_pb['City'] == "Delhi")) | (df_pb['Salary_INR'] > 70000)]
```

**ðŸ”¹ Using between**

```python
# Filter the DataFrame to include only rows where the 'Age' column value 
# is between 25 and 40 (inclusive by default).

filtered_df = df_pb[df_pb["Age"].between(25, 40)]
print(filtered_df)
```

```python
# Includes 25 but excludes 40
filtered_left = df_pb[df_pb["Age"].between(25, 40, inclusive="left")]
print("Includes 25:\n",filtered_left)

# Includes 40 but excludes 25
filtered_right = df_pb[df_pb["Age"].between(25, 40, inclusive="right")]
print("\nIncludes 40:\n",filtered_right)

# Excludes both 25 and 40
filtered_neither = df_pb[df_pb["Age"].between(25, 40, inclusive="neither")]
print("\nExcludes both 25 and 40:\n",filtered_neither)
```

```python
# ----------------------------------------------
# ðŸ“˜ PANDAS .query() METHOD â€” COMPLETE REFERENCE
# ----------------------------------------------

import pandas as pd

# Sample DataFrame for demonstration
data = {
    'Name': ['Amit', 'Neha', 'Raj', 'Priya', 'Karan', 'Anita'],
    'Age': [22, 27, 35, 29, 40, 25],
    'City': ['Delhi', 'Mumbai', 'Delhi', 'Pune', 'Delhi', 'Chennai'],
    'Salary': [40000, 52000, 61000, 48000, 72000, 45000]
}
df_pb = pd.DataFrame(data)

print("ðŸ”¹ Original DataFrame:\n", df_pb, "\n")

# ----------------------------------------------
# 1ï¸âƒ£ Basic Filtering
# ----------------------------------------------

print("1ï¸âƒ£ Basic Filtering â†’ Age > 25")
print(df_pb.query("Age > 25"), "\n")

# ----------------------------------------------
# 2ï¸âƒ£ Multiple Conditions (AND / OR)
# ----------------------------------------------

print("2ï¸âƒ£ AND condition â†’ Age > 25 and City == 'Delhi'")
print(df_pb.query("Age > 25 and City == 'Delhi'"), "\n")

print("   OR condition â†’ City == 'Delhi' or City == 'Mumbai'")
print(df_pb.query("City == 'Delhi' or City == 'Mumbai'"), "\n")

print("   Combined logical condition â†’ (Age > 25 and City == 'Delhi') or (Salary > 60000)")
print(df_pb.query("(Age > 25 and City == 'Delhi') or (Salary > 60000)"), "\n")

# ----------------------------------------------
# 3ï¸âƒ£ Using Variables Inside Query
# ----------------------------------------------

min_age = 25
target_city = "Delhi"

print("3ï¸âƒ£ Using Variables â†’ Age > @min_age and City == @target_city")
print(df_pb.query("Age > @min_age and City == @target_city"), "\n")

# ----------------------------------------------
# 4ï¸âƒ£ Filtering with Lists (using `in` / `not in`)
# ----------------------------------------------

print("4ï¸âƒ£ City IN ['Delhi', 'Mumbai']")
print(df_pb.query("City in ['Delhi', 'Mumbai']"), "\n")

print("   City NOT IN ['Delhi', 'Mumbai']")
print(df_pb.query("City not in ['Delhi', 'Mumbai']"), "\n")

# ----------------------------------------------
# 5ï¸âƒ£ Filtering on Numeric Ranges
# ----------------------------------------------

print("5ï¸âƒ£ Age BETWEEN 25 and 35 â†’ (25 <= Age <= 35)")
print(df_pb.query("25 <= Age <= 35"), "\n")

# ----------------------------------------------
# 6ï¸âƒ£ Filtering Using Expressions
# ----------------------------------------------

print("6ï¸âƒ£ Salary > Age * 2000")
print(df_pb.query("Salary > Age * 2000"), "\n")

# ----------------------------------------------
# 7ï¸âƒ£ Combining Query Results with .loc or Variables
# ----------------------------------------------

print("7ï¸âƒ£ Store query result in a variable (Age > 25 and City == 'Delhi')")
filtered_df = df_pb.query("Age > 25 and City == 'Delhi'")
print(filtered_df, "\n")

print("   Use .query() results with .loc for further filtering (Salary > 50000)")
print(df_pb.loc[df_pb.query("Salary > 50000").index, ["Name", "City", "Salary"]], "\n")

# ----------------------------------------------
# 8ï¸âƒ£ Querying with String Columns and Logical NOT
# ----------------------------------------------

print("8ï¸âƒ£ City != 'Delhi'")
print(df_pb.query("City != 'Delhi'"), "\n")

# ----------------------------------------------
# 9ï¸âƒ£ Querying with Complex Expressions
# ----------------------------------------------

print("9ï¸âƒ£ Age ** 2 > 800")
print(df_pb.query("Age ** 2 > 800"), "\n")

print("   (Salary / Age) > 1500 and City == 'Delhi'")
print(df_pb.query("(Salary / Age) > 1500 and City == 'Delhi'"), "\n")

# ----------------------------------------------
# ðŸ”Ÿ Chaining Query with Other Operations
# ----------------------------------------------

print("ðŸ”Ÿ Filter â†’ Sort â†’ Select Columns")
result = (
    df_pb.query("City == 'Delhi' and Age > 25")[["Name", "Age", "Salary"]]
    .sort_values("Salary", ascending=False)
)
print(result, "\n")

# ----------------------------------------------
# âœ… Summary
# ----------------------------------------------
# .query() is powerful for:
# - Readable and SQL-like filtering
# - Dynamic conditions with variables (@var)
# - Handling multiple logical operations cleanly
# - Reducing parentheses clutter compared to boolean indexing
```
